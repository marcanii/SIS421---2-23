```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
# link de datasets https://www.kaggle.com/datasets/gauravduttakiit/water-potability-prediction?select=train_dataset.csv

data = pd.read_csv('train_dataset.csv')
data
```

```python
# Agrega una nueva columna llamada 'day' con valores del 1 al nÃºmero de filas
num_filas = len(data)
data.insert(0, 'day', range(1, num_filas + 1))

# Guarda el DataFrame modificado de nuevo en un archivo CSV
nuevo_archivo_csv = 'nuevo_archivo.csv'
data.to_csv(nuevo_archivo_csv, index=False)

print("Columna 'day' agregada al principio exitosamente.")
data
```

```python
data = np.loadtxt('nuevo_archivo.csv', delimiter=',', skiprows=1)
X, y = data[:, 0:10], data[:, 10]
print(X.shape)
print(y.shape)
```

```python

num_features = 10
val_min = np.min(X)
val_max = np.max(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Generar pesos directamente proporcionales a los valores de las caracterÃ­sticas en el rango [0, 700]
weights = np.linspace(val_min, val_max, num_features)

# Calcular la puntuaciÃ³n transformada Y con pesos proporcionales
Y_transformed = np.dot(X_scaled, weights)

# Normalizar los valores transformados de Y al rango [0, 1]
Y_transformed = (Y_transformed - np.min(Y_transformed)) / (np.max(Y_transformed) - np.min(Y_transformed))
Y_transformed = np.round(Y_transformed, decimals=5)

print("X_scaled.shape:", X_scaled.shape)
print("Y_transformed.shape:", Y_transformed.shape)
print("Y_transformed:", Y_transformed)
```

```python
# Combina las matrices X y Y en una sola matriz
data = np.column_stack((X, Y_transformed))

# Guarda la matriz en un archivo CSV
nombre_archivo = 'water_quality_train.csv'
np.savetxt(nombre_archivo, data, delimiter=',', header="day,ph,Hardness,Solids,Chloramines,Sulfate,Conductivity,Organic_carbon,Trihalomethanes,Turbidity,Probability", comments="")

print("Datos guardados en", nombre_archivo)
```
